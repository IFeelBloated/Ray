```python
import matplotlib.pyplot as plt
from matplotlib.collections import EventCollection
import numpy as np
```

```python
np.arange(0, 1, 0.01, dtype=float)
```

```python
# split the data into two parts
x = np.arange(0, 1, 0.01, dtype=float)
b = 1

y1 = 1 - np.sin(x/b*np.pi/2)
y2 = 1 - x/b
y3 = 1 - ( -2 * (x/b)**3 + 3 * (x/b)**2)
y4 = 1 - (np.cos(x/b*np.pi) + 1)/2


# plot the data
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(x, y1, color='tab:blue')
ax.plot(x, y2, color='tab:red')
ax.plot(x, y3, color='tab:green')
ax.plot(x, y4, color='yellow')

# set the limits
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])

ax.set_title('line plot with data points')

# display the plot
plt.show()
```
